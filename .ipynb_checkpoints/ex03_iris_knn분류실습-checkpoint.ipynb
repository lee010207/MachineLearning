```python
iris_df = pd.DataFrame(iris_data['data'], columns=iris_data['feature_names'])
iris_df.head()
```

```python
X = iris_df

# target  
y = iris_data.target
print(X.shape, y.shape)
```

```python
# ,  
# train_test_split(, ,  ) 
# : (1)  , (2) ,  
#  7:3 ( )
# train  : 0.7 / test  : 0.3 (30%)
# 7.5:2.5 , 8:2   
#    ,
#   

X_train,X_test,y_train,y_test = ['','','','']
X_train,X_test,y_train,y_test = train_test_split(X , y , test_size=0.25, random_state = 1)
# random_state =  :  ()
# random_state shuffle   

X_train

# Xy_data = train_test_split(X , y , test_size=0.3)
# X_train = Xy_data[0]
# X_test = Xy_data[1]
# y_train = Xy_data[2]
# y_test = Xy_data[3]
```

```python
#     5 
#  iris_knn
iris_knn = KNeighborsClassifier(n_neighbors=5)
```

```python
#  
# .fit(, )
iris_knn.fit(X_train, y_train)
```

```python
#  
# pre  
# .predict
pre = iris_knn.predict(X_test)
pre
```

```python
accuracy_score(y_test, pre)
# 7.5:2.5    knn    ?
#   1 
#    1 :  ()
```

```python
iris_data.target_names[pre]
```

-  
  - n_neighbors :   ->     !

```python
#    2 
# 1,3,5,7,9,..      
#  
train_lst = []
test_lst = []

# 1 49 2   
n_setting = range(1,100,2)   #  

for i in n_setting:
    #  
    knn_clf = KNeighborsClassifier(n_neighbors=i)
    #   train  
    knn_clf.fit(X_train, y_train)
    #    ( )
    # train -> train_lst 
    train_pre = knn_clf.predict(X_train)
    train_acc = accuracy_score(y_train,train_pre)
    train_lst.append(train_acc)
    # test -> test_lst 
    test_pre = knn_clf.predict(X_test)
    test_acc = accuracy_score(y_test,test_pre)
    test_lst.append(test_acc)
```

```python
for i in train_lst:
    print(i)
```

```python
for i in test_lst:
    print(i)
```

```python
# k     
#  
#    
plt.figure(figsize = (10,4))
plt.plot(n_setting,train_lst,label='train acc')
plt.plot(n_setting, test_lst, label='test acc')
plt.legend()
plt.grid()
plt.ylabel("accuracy")   # y  
plt.xlabel("n_neighbors==i")    # x  
plt.xticks(n_setting)       # x   
plt.show()
```

```python
# iris_knn   
#    
from sklearn.model_selection import cross_val_score
```

```python
# , ,  
# cross_val_score(, train , cv=)
# cv :    (  !)
# k=5(n_neighbors=5) ( 5)
cross_val_score(iris_knn , X_train,y_train, cv=10).mean()
```

```python
#  fit()     
# k=9
iris_knn9 = KNeighborsClassifier(n_neighbors=9)
print(cross_val_score(iris_knn9 , X_train,y_train, cv=10))
print("  : ",cross_val_score(iris_knn9 , X_train,y_train, cv=10).mean())

# k=3
iris_knn3 = KNeighborsClassifier(n_neighbors=3)
print(cross_val_score(iris_knn3 , X_train,y_train, cv=10))
print("  : ",cross_val_score(iris_knn3 , X_train,y_train, cv=10).mean())

#        
#  
# -> 9   
```
