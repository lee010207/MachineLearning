### 데이터 정리
- 문제 데이터를 DataFrame 형태로 변환 (2차원)
- 문제 데이터와 답데이터로 변수 정리(X,y)
- train, test 분리 (train_test_split 활용)
  - X_train , X_test, y_train, y_test 

```python
iris_df = pd.DataFrame(iris_data['data'], columns=iris_data['feature_names'])
iris_df.head()
```

```python
X = iris_df

# target 데이터 연결하기
y = iris_data.target
print(X.shape, y.shape)
```

```python
# 훈련, 테스트 분리
# train_test_split(문제, 답, 테스트 사이즈) 
# : (1) 랜덤 섞어줌, (2) 훈련, 테스트 분리
# 통상적으로 7:3으로 분리(정답은 아님)
# train 크기 : 0.7 / test 크기 : 0.3 (30%)
# 7.5:2.5 , 8:2 로 분리하기도 함
# 비율적인 의미로 넣어주려면 실수형태,
# 개수의미로 넣어주려면 정수형태

X_train,X_test,y_train,y_test = ['훈련문제','테스트문제','훈련답','테스트답']
X_train,X_test,y_train,y_test = train_test_split(X , y , test_size=0.25, random_state = 1)
# random_state = 정수형태 : 랜덤규칙 조정(고정시키려고)
# random_state shuffle을 동일한 기준으로 실행

X_train

# Xy_data = train_test_split(X , y , test_size=0.3)
# X_train = Xy_data[0]
# X_test = Xy_data[1]
# y_train = Xy_data[2]
# y_test = Xy_data[3]
```

```python
print(f'훈련용 문제 데이터 : {X_train.shape} , 테스트용 문제:{X_test.shape}')
print(f'훈련용 답 : {len(y_train)},테스트용 답: {y_test.shape}')
# 문제개수 만큼 답 개수가 셋팅
# 훈련용 특성의 개수만큼 테스트용 특성 개수가 셋팅
```

```python
#분리된 데이터 클래스별 개수 확인하기
#7:3 비율 데이터 분리, 모델 입장에서 학습을 잘 하려면 다양한 데이터가 필요함
pd.Series(y_train).value_counts()
np.unique(y_train, return_counts = True)[1]/112
# array([0.34285714, 0.3047619 , 0.35238095]) -> 7:3
# array([0.33035714, 0.30357143, 0.36607143]) -> 7.5/2.5
# 비율을 바꿔봤을 때 다양성이 유지되는 쪽으로 설정값을 사용!
```

### 모델링
- 모델 객체 생성 및 하이퍼 파라미터 조정
- 모델 학습
- 모델 예측 및 평가

```python
# 객체 생성 이웃의 개수 5로 설정
# 모델명 iris_knn
iris_knn = KNeighborsClassifier(n_neighbors=5)
```

```python
# 모델 학습
# 모델.fit(훈련문제, 훈련답)
iris_knn.fit(X_train, y_train)
```

```python
# 모델 예측
# pre 변수에 담아보기
# 모델.predict
pre = iris_knn.predict(X_test)
pre
```

```python
accuracy_score(y_test, pre)
# 7.5:2.5로 분리한 데이터를 학습한 knn모델을 바로 믿을 수 있는가?
# 평가를 딱 1번만 진행
# 새롭게 예측한 데이터는 1번 : 여러번 검증(교차검증)
```

```python
iris_data.target_names[pre]
```

- 하이퍼 파라미터
  - n_neighbors : 이웃 수 -> 여러번 바꿔보면서 모델의 성능을 확인해보자!

```python
# 저장하는 빈 리스트 2개 설정
# 1,3,5,7,9,.. 홀수로 이웃수를 바꿔가면서 모델 성능을 확인해볼거야
# 반복문 활용
train_lst = []
test_lst = []

# 1부터 49까지 2칸씩 뛰어서 홀수값을 담음
n_setting = range(1,100,2)   # 이웃의 수

for i in n_setting:
    # 모델 생성
    knn_clf = KNeighborsClassifier(n_neighbors=i)
    # 모델 학습 train 데이터 활용
    knn_clf.fit(X_train, y_train)
    # 모델 예측 및 평가(성능 확인)
    # train -> train_lst 저장
    train_pre = knn_clf.predict(X_train)
    train_acc = accuracy_score(y_train,train_pre)
    train_lst.append(train_acc)
    # test -> test_lst 저장
    test_pre = knn_clf.predict(X_test)
    test_acc = accuracy_score(y_test,test_pre)
    test_lst.append(test_acc)
```

```python
for i in train_lst:
    print(i)
```

```python
for i in test_lst:
    print(i)
```

```python
# k값이 바뀜에 따른 성능 추이 확인
# 선 그래프
#하이퍼 파라미터 튜닝 결과 시각화
plt.figure(figsize = (10,4))
plt.plot(n_setting,train_lst,label='train acc')
plt.plot(n_setting, test_lst, label='test acc')
plt.legend()
plt.grid()
plt.ylabel("accuracy")   # y축 이름 설정
plt.xlabel("n_neighbors==i")    # x축 이름 설정
plt.xticks(n_setting)       # x축 격자 크기 설정
plt.show()
```

```python
# train 100% 맞춘다는 건 과대적합일 확률이 높을 수 있음
# i가 높아질수록 성능이 떨어지고 있음
# -> 점점 과소적합일 확률이 높아지고 있음
# 학습용 데이터라서 모델이 학습하기에, 예측하기에 쉬운 데이터
# 데이터에 대해서 여러번 검증해서 확인해봐야 함 (교차검증)
# -> 신뢰 가능한 모델인지 판단
```

```python
# iris_knn 모델 교차검증 진행
# 모델의 일반화 성능 확인
from sklearn.model_selection import cross_val_score
```

```python
# 모델, 데이터, 몇번 검증
# cross_val_score(모델, train 데이터, cv=횟수)
# cv : 데이터를 분리해서 검증하는 횟수(몇번 정확도를 계산하겠다!)
# k=5(n_neighbors=5)일 때(이웃수가 5)
cross_val_score(iris_knn , X_train,y_train, cv=10).mean()
```

```python
# 교차검증은 fit()을 하지 않아도 정확도 측정 가능
# k=9
iris_knn9 = KNeighborsClassifier(n_neighbors=9)
print(cross_val_score(iris_knn9 , X_train,y_train, cv=10))
print("교차검증 평균 : ",cross_val_score(iris_knn9 , X_train,y_train, cv=10).mean())

# k=3
iris_knn3 = KNeighborsClassifier(n_neighbors=3)
print(cross_val_score(iris_knn3 , X_train,y_train, cv=10))
print("교차검증 평균 : ",cross_val_score(iris_knn3 , X_train,y_train, cv=10).mean())

# 교차검증의 결과를 확인하면서 어떤 파라미터로 설정하는게 좋을지 확인
# 사람이 판단함
# -> 9로 설정하는 것도 괜찮음
```
