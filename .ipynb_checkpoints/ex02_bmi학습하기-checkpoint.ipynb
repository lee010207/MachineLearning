```python
data = data.reset_index('Label')
```

```python
X = data.loc[:,'Height':'Weight']
y = data['Label']
```

```python
X_train = X.iloc[:350,:]
X_test = X.iloc[350:,:]
X_test
```

```python
y_train = y[:350]
# y.iolc[:350]

y_test = y[350:]
# y.iloc[350:]
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

### 모델링
- 모델 선택 및 하이퍼 파라미터 조정(knn모델 선택)
- 모델 객체 생성
- 모델 학습
- 모델 예측 및 평가

```python
# 데이터 클래스(카테고리)의 개수? 6
from sklearn.neighbors import KNeighborsClassifier
#여섯개 중 한개를 판단하는 것 : 다중분류
#정확도 점수(정확도 지표)
from sklearn.metrics import accuracy_score
```

```python
# knn_model 이라는 변수로 모델 객체 생성
# 사람이 설정하는 매개변수 : 하이퍼 파라미터
# 이웃의 수 : n_neighbors
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model
```

```python
# 모델 학습
# fit(훈련용 문제 데이터, 훈련용 답 데이터)
knn_model.fit(X_train, y_train)
```

```python
# 에측 : 학습했나 확인
# predict(테스트용 문제)
pre = knn_model.predict(X_test)
pre
```

```python
pre.size
```

```python
# 평가
# 정확도 지표
# accuracy_score(테스트용 답 데이터, 예측한 값)
accuracy_score(y_test, pre)
# 90.6666666...%(약 91%)
```
